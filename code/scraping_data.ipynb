```python
years = list(range(2000,2020,1))
def get_stats(years):

    url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
    
    # create list of dataframes 
    year_df_list = []
    
    # loops through all url links, formats the link with the given year, creates a dataframe.
    for year in years:
        url = url_link.format(year)
        df = pd.read_html(url, header = 0)
        df = pd.DataFrame(df[0])
        year_df_list.append(df)
    # Concat all dataframes in list
    trad_stats = pd.concat(year_df_list, ignore_index=True)
        
    return trad_stats
trad_stats = get_stats(years)
```

```python
# Dropping duplicate Headers 
trad_stats = trad_stats.drop(trad_stats[trad_stats.Age == 'Age'].index)
trad_stats = trad_stats.drop(columns='Rk')
trad_stats.head()
```

```python
int_list = ['Age','GS','G','Player','Tm','Pos']
```

```python
# Converting dtypes in dataframe

for feature in trad_stats:
    if feature not in int_list:
        trad_stats[feature] = trad_stats[feature].astype(float)


trad_stats['Age'] = trad_stats['Age'].astype(int)
trad_stats['GS'] = trad_stats['GS'].astype(int)
trad_stats['G'] = trad_stats['G'].astype(int)

trad_stats.info()
```
