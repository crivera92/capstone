## ```Pulling Advanced Stats```

````python
driver = webdriver.Chrome(executable_path="/Users/crivera/Desktop/chromedriver")

url = r'https://stats.nba.com/players/advanced/?sort=PTS&dir=-1'
driver.get(url)

# Declaring 
select = Select(driver.find_element_by_xpath(r"/html/body/main/div/div/div[2]/div/div/nba-stat-table/div[1]/div/div/select"))

select.select_by_index(0)

src = driver.page_source
parser = BeautifulSoup(src, "lxml")
table = parser.find('div', attrs = {"class":"nba-stat-table__overflow"})
headers = table.findAll('th')
headerslist = [h.text.strip() for h in headers[1:]]

headerslist

# Removing all rank headers from list
headerslist1 = [a for a in headerslist if not 'RANK' in a]
headerslist1

rows = table.findAll('tr')[1:]

len(rows)

player_stats = [[td.getText().strip() for td in rows[i].findAll('td')[1:]] for i in range(len(rows))]
player_stats[0]

players_stats = pd.DataFrame(player_stats, columns=headerslist1[:-5])

players_stats
````

````python
# Converting dtypes in dataframe

int_list = ['AGE','GP','W','L','TEAM','Player','POSS']

for feature in nba_stats2021.columns:
    if feature not in int_list:
        nba_stats2021[feature] = nba_stats2021[feature].astype(float)
        
for feature in int_list[:-3]:
    print(feature)
    nba_stats2021[feature] = nba_stats2021[feature].astype(int)

nba_stats2021.info()
````

````python
# Funciton to remove comma from number of POSS (possesions)
# before converting the dtypes to integer

def remove_comma(column):
    num = column.split(',')
    num = ''.join(num)
    return num

nba_stats2021['POSS'] = nba_stats2021['POSS'].apply(remove_comma)
nba_stats2021['POSS'] = nba_stats2021['POSS'].astype(int)
````
