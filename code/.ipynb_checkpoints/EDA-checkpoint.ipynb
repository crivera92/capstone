```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
nba_stats2021 = pd.read_csv('nba_stats2021_original.csv')
nba_stats2021.head()
```

```python
nba_stats2021.describe()
```

```python
nba_stats2021.shape
```

```python
nba_stats2021.columns
```

```python
# Creating List of offensive stat features
off_stat = []
for feature in nba_stats2021.columns:
    if feature.startswith(('3','2','A','PT','%P','TO','e','PO')):
        off_stat.append(feature)
print(len(off_stat))
off_stat
```

```python
# Creating List of defensive stat features
def_stat = []
for feature in nba_stats2021.columns:
    if feature.startswith(('D','%D','S','%S','B','R','%B')):
        def_stat.append(feature)
print(len(def_stat))
def_stat
```

```python
figure(figsize=(10,6),dpi=80)
sns.heatmap(nba_stats2021[def_stat].corr(), annot=True);
```

```python
figure(figsize=(15,6), dpi=80)
sns.heatmap(nba_stats2021[off_stat].corr(), annot=True);
```

```python
sns.displot(nba_stats2021['MIN'], kde=False);
# Checking what the average amount of minutes a player plays
print(f"Average Minutes Played: {nba_stats2021['MIN'].mean()}")
```

```python
sns.boxplot(nba_stats2021['MIN']);
plt.title('Distribution of Minutes PLayed');
```

```python
sns.displot(nba_stats2021['PTS'], kde=False);
# Checking average amount of points scored by a player. 
print(f"Average points scored: {nba_stats2021['PTS'].mean()}")
```

```python
sns.boxplot(nba_stats2021['PTS']);
plt.title('Distribution of Points Scored');
```

```python
sns.displot(nba_stats2021['GP'])
plt.title('Average Games Played by PLayer')
print(f"Average games played by player: {nba_stats2021['GP'].mean()}")
```

```python
sns.boxplot(nba_stats2021['GP'])
plt.title('Distribution of Games PLayed');
```
